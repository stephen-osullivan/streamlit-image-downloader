```python
import matplotlib.pyplot as plt
import math
import os
from PIL import Image
# view the images
files = os.listdir('downloads',)
fig, axs = plt.subplots(math.ceil(len(files)/2),2, figsize=(20,15))
for file, ax in zip(files, axs.ravel()):
    ax.imshow(Image.open(os.path.join('downloads', file)))
    if len(file) > 50:
        # truncate file name if necessary
        file = "... " + file[-50:]
    ax.set_title(file)
plt.tight_layout()
plt.show()
```
